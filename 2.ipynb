1. Download from yahoo the following tickers = ['META', 'GOOG', 'AMZN', 'SQ', 'TSLA', '^GSPC'], from 2020/01/01.

```python
!pip install yfinance --quiet
import yfinance as yf

Meta = yf.download('META',
                       start = '2020-01-01',
                       #end = '2023-08-15',
                       progress = False
                       )

Google = yf.download('GOOG',
                       start = '2020-01-01',
                       #end = '2023-08-15',
                       progress = False
                       )

Amazon = yf.download('AMZN',
                       start = '2020-01-01',
                       #end = '2023-08-15',
                       progress = False
                       )

Block = yf.download('SQ',
                       start = '2020-01-01',
                       #end = '2023-08-15',
                       progress = False
                       )

Tesla = yf.download('TSLA',
                       start = '2020-01-01',
                       #end = '2023-08-15',
                       progress = False
                       )

SP500 = yf.download('^GSPC',
                       start = '2020-01-01',
                       #end = '2023-08-15',
                       progress = False
                       )
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
Meta.head()
```

2. Plot the "Adj Close" for all equties in the very same plot using Plotly.

```python
trace1 = go.Scatter(x= Meta.index, y=Meta['Adj Close'], mode='lines', name='Meta')
trace2 = go.Scatter(x= Google.index, y=Google['Adj Close'], mode='lines', name='Google')
trace3 = go.Scatter(x= Amazon.index, y=Amazon['Adj Close'], mode='lines', name='Amazon')
trace4 = go.Scatter(x= Block.index, y=Block['Adj Close'], mode='lines', name='Block')
trace5 = go.Scatter(x= Tesla.index, y=Tesla['Adj Close'], mode='lines', name='Tesla')
trace6 = go.Scatter(x= SP500.index, y=SP500['Adj Close'], mode='lines', name='S&P500 Index')


# Create a layout
layout = go.Layout(
    title='S&P 500 Index and Adjusted Close Price for 5 Stocks since 2020-01-01',
    xaxis=dict(title='Time'),
    yaxis=dict(title='Price')
)

# Create a figure and add traces
fig = go.Figure(data=[trace1, trace2, trace3, trace4, trace5, trace6], layout=layout)

# Show the plot
fig.show()
```

3. Calculte the precentage of value increase for all the companies, or Standarize Price from 2020/01/01.
4. Plot the Standarize plot in a Plotly for all tickers

```python
Meta['Standardized Closing Price']=(Meta['Close'] - Meta['Close'].iloc[0]) / Meta['Close'].iloc[0] * 100
Google['Standardized Closing Price']=(Google['Close'] - Google['Close'].iloc[0]) / Google['Close'].iloc[0] * 100
Amazon['Standardized Closing Price']=(Amazon['Close'] - Amazon['Close'].iloc[0]) / Amazon['Close'].iloc[0] * 100
Block['Standardized Closing Price']=(Block['Close'] - Block['Close'].iloc[0]) / Block['Close'].iloc[0] * 100
Tesla['Standardized Closing Price']=(Tesla['Close'] - Tesla['Close'].iloc[0]) / Tesla['Close'].iloc[0] * 100
SP500['Standardized Closing Price']=(SP500['Close'] - SP500['Close'].iloc[0]) / SP500['Close'].iloc[0] * 100
```

```python
trace1 = go.Scatter(x= Meta.index, y=Meta['Standardized Closing Price'], mode='lines', name='Meta')
trace2 = go.Scatter(x= Google.index, y=Google['Standardized Closing Price'], mode='lines', name='Google')
trace3 = go.Scatter(x= Amazon.index, y=Amazon['Standardized Closing Price'], mode='lines', name='Amazon')
trace4 = go.Scatter(x= Block.index, y=Block['Standardized Closing Price'], mode='lines', name='Block')
trace5 = go.Scatter(x= Tesla.index, y=Tesla['Standardized Closing Price'], mode='lines', name='Tesla')
trace6 = go.Scatter(x= SP500.index, y=SP500['Standardized Closing Price'], mode='lines', name='S&P500 Index')


# Create a layout
layout = go.Layout(
    title='S&P 500 Index and Standardized Closing Price for 5 Stocks since 2020-01-01',
    xaxis=dict(title='Time'),
    yaxis=dict(title='Price')
)

# Create a figure and add traces
fig = go.Figure(data=[trace1, trace2, trace3, trace4, trace5, trace6], layout=layout)

# Show the plot
fig.show()
```

5. Calculate current increase since 2020/01/01, and show it in the command window.

```python
#Meta, Google, Amazon, Block, Tesla, S&P 500
Meta_last = Meta['Standardized Closing Price'].iloc[-1]
Google_last = Google['Standardized Closing Price'].iloc[-1]
Amazon_last = Amazon['Standardized Closing Price'].iloc[-1]
Block_last = Block['Standardized Closing Price'].iloc[-1]
Tesla_last = Tesla['Standardized Closing Price'].iloc[-1]
SP500_last = SP500['Standardized Closing Price'].iloc[-1]

print('Meta current price % increase compared to 2020-01-01: ', Meta_last)
print('Google current price % increase compared to 2020-01-01: ', Google_last)
print('Amazon current price % increase compared to 2020-01-01: ', Amazon_last)
print('Block current price % increase compared to 2020-01-01: ', Block_last)
print('Tesla current price % increase compared to 2020-01-01: ', Tesla_last)
print('S&P 500 current index % increase compared to 2020-01-01: ', SP500_last)
```
